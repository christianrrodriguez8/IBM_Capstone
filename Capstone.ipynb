### Problem Description  -- Features
This Capstone project will use a Data Science approach to predict the likelihood that a new restaurant opened in London, UK will be successful. To train this predictive model, features of existing restaurants to be considered will include location, cuisine type (e.g. Indian or Italian), and menu pricing. Once trained, the model can be used to adjust these features for a proposed restaurant (e.g. changing menu pricing or moving to a different location), optimizing its chances for success. Once trained, some example predictions will be made as a proof-of-concept (for example, if we wish to open an Italian restaurant, and we insist on making it an upscale restaurant with higher menu prices, where would be the best locations in London to choose for this restaurant?).

### Data Description -- Quantifying Success
In addition to the features that will need to be obtained for each London restaurant (location, cuisine, pricing) from the Foursquare API, properties that serve to quantify the success of each restaurant must be obtained. In an ideal case, the success of a restaurant would be measured by its customer satisfaction and its revenue. Although customer satisfaction can be directly measured by Foursquare's rating system, it is not possible in general to directly determine how profitable a restaurant is. However, a restaurant's profitability can be inferred from two pieces of data that are available on Foursquare: the total number of tips on its page, and the local density of nearby restaurants. The number of tips on a restaurant's page can be expected to be proportional to its foot traffic, and the density of nearby restaurants determines if the restaurant is in a "hot spot" where many restaurant-goers are expected to be found. A single "success score" will be computed for each restaurant as a function of these three properties (rating, tip count, and density), and the values that determine the relative importance of each property will be hyperparameters that can be tuned through successive iterations of training and testing the model.
